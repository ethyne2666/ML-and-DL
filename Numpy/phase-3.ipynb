```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("==== Zomato sales analysis ==== ")
print("\n Sales data shape", sales_data.shape)
print("\n Sample data for 1st 3 restau: ", sales_data[: ,1:])
```

```python
# total sales per year

print(np.sum(sales_data[:, 1:], axis=0))  # Sum across rows for each year

yearly_total = np.sum(sales_data[:, 1:], axis=0)
print("\n Yearly total sales from 2021 to 2024: ", yearly_total)


#Minimum sales per restaurant
min_sales = np.min(sales_data[:, 1:], axis=1)
print("\n Minimum sales per restaurant from 2021 to 2024: ", min_sales)

# Maximum sales per year
max_sales = np.max(sales_data[:, 1:], axis=0)
print("\n Maximum sales per year from 2021 to 2024: ", max_sales)

# Average sales per restautant
avg_sales = np.mean(sales_data[:, 1:], axis=1)
print("\n Average sales per restaurant from 2021 to 2024: ", avg_sales)
```

```python
cumsum = np.cumsum(sales_data[:, 1:], axis=0)
print("\n Cumulative sales from 2021 to 2024: ", cumsum)

plt.figure(figsize=(8,6))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Average Cumulative Sales from 2021 to 2024")
plt.xlabel("Years") 
plt.ylabel("Average Sales")
plt.xticks(ticks=[0,1,2,3], labels=['2021', '2022', '2023', '2024'])
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1, 2, 3, 4, 5])
vector2 = np.array([6, 7, 8, 9, 10])

print("Vector addition: ", vector1 + vector2)

print("\n Multiplication vector", vector1 * vector2)

print("\nDot Product", np.dot(vector1, vector2))

angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print(angle)
```
